```python
# ------------------------------------------
# projet : données des produits de Place de marché
# Auteur : Rim BAHROUN
# Date: Fevrier 2023
# OpenClassrooms
# -------------------------------------------
# importation des librairies Python pour DS
# -------------------------------------------
import requests
import pandas as pd

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/parser"

querystring = {"ingr":"champagne"}

headers = {
    "X-RapidAPI-Key": "bd8c2f188cmshf5d0f1a1c8718c6p1c7f46jsn370366720478",
    "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

# Appel à l'API
response = requests.request("GET", url, headers=headers, params=querystring)

# Extraction des 10 premiers produits et création d'un dataframe
products = response.json()["hints"][:10]

# Extract required data from each product
data = []
for product in products:
    food = product['food']
    data.append({
        'foodId': food['foodId'],
        'label': food['label'],
        'category': food['category'],
        'foodContentsLabel': food.get('foodContentsLabel', ''),
        'image': food.get('image', '')
    })

# Create a pandas dataframe from the extracted data
df = pd.DataFrame(data, columns=['foodId', 'label', 'category', 'foodContentsLabel', 'image'])

# Save the dataframe to a CSV file
df.to_csv('products.csv', index=False)
```

Voici les cinq grands principes du RGPD :

Transparence et licéité : les entreprises doivent informer les personnes de manière transparente et claire sur la manière dont leurs données personnelles sont collectées, utilisées et traitées, et ne peuvent le faire que de manière licite, c'est-à-dire avec un fondement juridique approprié.

Finalité : les entreprises ne peuvent collecter et traiter des données personnelles que pour des finalités spécifiques et légitimes, qui doivent être clairement définies avant la collecte des données.

Minimisation des données : les entreprises doivent limiter la collecte et le traitement des données personnelles aux données qui sont nécessaires pour atteindre la finalité spécifique pour laquelle elles ont été collectées.

Exactitude : les entreprises doivent prendre des mesures pour s'assurer que les données personnelles qu'elles collectent et traitent sont exactes et à jour, et doivent permettre aux personnes de demander la correction de toute donnée inexacte.

Sécurité : les entreprises doivent prendre des mesures techniques et organisationnelles appropriées pour protéger les données personnelles contre la perte, le vol, la divulgation non autorisée et l'accès non autorisé, et doivent notifier les personnes en cas de violation de données susceptible de présenter un risque élevé pour leurs droits et libertés.
